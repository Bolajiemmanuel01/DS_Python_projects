```python
import pandas as pd
import numpy as np
stock = pd.read_csv("stock.csv")
stock
```

```python
vending_machine = pd.read_csv("vending_machine.csv")
vending_machine
```

```python
newsnacks = {"coca cola's coke":"Coke", "coca-cola's diet coke":"Diet Coke", "a.g. bar's irn-bru":"Irn-bru", 
             "Forrest Mars' mars":"Mars", "mars inc's milky way":"Milky Way", "mackies' sea salt crisps":"Sea Salt Crisps", 
             "walkers' shortbread":"Shortbread", "pepsico's doritos":"Doritos"}
data = stock.replace(newsnacks)
data
```

```python
joint_table = pd.merge(vending_machine, data, how="outer")
joint_table = joint_table.set_index('machine_code')
joint_table = joint_table.sort_index(ascending=True)
joint_table
```

```python
joint_table.loc['B4'] = ('Irn-bru', 3, 'drinks', 3.00)
joint_table
```

```python
while True:
    choice = input("Welcome to the vending machine. Select an option:\n1. Log in (admin only)\n"
                    "2. Add products (admin only)\n3. Buy products\n4. Exit\n>> ")
    if choice == '1':
        no_of_try = 1
        while no_of_try <= 3:
            user_name = input("Please enter your user name: ").lower()
            admins = ("bolaji", "emmanuel")
            if user_name not in admins:
                print(f"Invalid username, current attempts are {3 - no_of_try}")
                no_of_try += 1
            else:
                no_of_try = 1
                while no_of_try <= 3:
                    try:
                        password = int(input("Please Enter your Password: "))
                        if password == 2010:
                            print("great")
                        else:
                            print(f"invalid password , current attempts are {3 - no_of_try}")
                            no_of_try += 1
                    except:
                        print(f"password should be a number, current attempts are {3 - no_of_try}")
                        no_of_try += 1
        print("Exceeded Number of Try(s)")
        break

    elif choice == '2':
        print("Please Login First!")
    elif choice == '3':
        print("This is the current stock of the vending machine: ")
        print(joint_table)
        while True:
            product_ID = input("Select a product to buy: ").upper()
            try:
                product_details = joint_table.loc[product_ID]
                if pd.isnull(product_details['snack']):
                    print('No products left on this slot, try another one')
                else:
                    if product_details['amount'] == 0:
                        print('Product is currently empty, try another one')
                    else:
                        print(f'Product to buy: {product_details["snack"]}\nPrice: {product_details["price"]}')
                        while True:
                            try:
                                payment = input('Please enter the amount you want to pay: ')
                                payment = float(payment)
                                if payment <= 0:
                                    print("you inserted an invalid aount! try again!")
                                else:
                                    if payment >= product_details['price']:
                                        change = payment - product_details['price']
                                        print('good')
                                        break
                                    else:
                                        print("you need to pay more, try again")
                            except:
                                print('You entered an invalid amount!')
                        
            except KeyError:
                print('Invalid slot, try again')
            except:
                print('Invalid product')
    elif choice == '4':
        print("Thank you for Shopping with us!")
        break
    else:
        print("Invalid Option. Try again!")
```
